```python
import pandas as pd

df = pd.read_csv(r"/Users/nieveenalia/Downloads/U.S._Chronic_Disease_Indicators.csv", low_memory=False)
print(sorted(df["Topic"].unique())[:30])  # See the first 30 topic names
```

```python
topics_keep = ["Tobacco", "Diabetes", "Nutrition, Physical Activity, and Weight Status"]
df_focus = df[df["Topic"].isin(topics_keep)]

print(df_focus["Topic"].value_counts())  # sanity check
df_focus.head()
```

```python
print(sorted(df_focus["YearStart"].unique()))
```

```python
# Check all years
print(sorted(df_focus["YearStart"].unique()))

# Keep only the last 7
last_7_years = sorted(df_focus["YearStart"].unique())[-7:]
df_focus = df_focus[df_focus["YearStart"].isin(last_7_years)]

print(sorted(df_focus["YearStart"].unique()))  # confirm




```

```python
df_focus = df_focus[["YearStart", "LocationDesc", "Topic", "Question", "DataValue"]]
print(df_focus.head())
```

```python
print(df_focus.isna().sum())  # count missing values

# Drop rows with missing DataValue (optional, but usually better)
df_focus = df_focus.dropna(subset=["DataValue"])
```

```python
df_focus.to_csv("chronic_disease_cleaned.csv", index=False)
```

```python
df_focus["DataValue"] = pd.to_numeric(df_focus["DataValue"], errors="coerce")
```

```python
# Average diabetes rates by year
print(
    df_focus[df_focus["Topic"] == "Diabetes"]
    .groupby("YearStart")["DataValue"]
    .mean()
)

# Top 5 states with highest tobacco use
print(
    df_focus[df_focus["Topic"] == "Tobacco"]
    .groupby("LocationDesc")["DataValue"]
    .mean()
    .sort_values(ascending=False)
    .head()
)
```

```python
df_diabetes = df_focus[df_focus["Topic"] == "Diabetes"]
```

```python
avg_diabetes = df_diabetes.groupby("LocationDesc")["DataValue"].mean()
```

```python
top5_diabetes = avg_diabetes.sort_values(ascending=False).head(5)
print(top5_diabetes)
```

```python

# 1. Filter for Tobacco only (correct spelling + lowercase variable)
df_tobacco = df_focus[df_focus["Topic"] == "Tobacco"]

# 2. Remove 'United States' since we only want states
df_tobacco = df_tobacco[df_tobacco["LocationDesc"] != "United States"]

# 3. Group by State and calculate average DataValue
avg_tobacco = df_tobacco.groupby("LocationDesc")["DataValue"].mean()

# 4. Sort descending and get top 5 states
top5_tobacco = avg_tobacco.sort_values(ascending=False).head(5)
print(top5_tobacco)
```

```python
# 1. Filter by Topic
df_nutrition = df_focus[df_focus["Topic"] == "Nutrition, Physical Activity, and Weight Status"]

# 2. Narrow down to obesity-related questions only
df_obesity = df_nutrition[df_nutrition["Question"].str.contains("obesity", case=False, na=False)]

# 3. Remove 'United States' since we only want states
df_obesity = df_obesity[df_obesity["LocationDesc"] != "United States"]

# 4. Group by State and calculate average DataValue
avg_obesity = df_obesity.groupby("LocationDesc")["DataValue"].mean()

# 5. Sort descending and get top 5 states
top5_obesity = avg_obesity.sort_values(ascending=False).head(5)
print(top5_obesity)
```

```python
df_focus = df_focus[df_focus["LocationDesc"] != "United States"]
```

```python
df_focus.groupby("Topic")["DataValue"].describe()
```

```python
df_focus = df_focus[(df_focus["DataValue"] >= 0) & (df_focus["DataValue"] <= 100)]
```

```python
avg_trends = (
    df_focus
    .groupby(["YearStart", "Topic"])["DataValue"]
    .mean()
    .reset_index()
)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.lineplot(data=avg_trends, x="YearStart", y="DataValue", hue="Topic", marker="o")
plt.title("Trends Over Time: Diabetes, Tobacco, and Obesity")
plt.ylabel("Average Prevalence (%)")
plt.show()
```

```python
max_diabetes = (
    avg_trends[avg_trends["Topic"] == "Diabetes"]
    .sort_values("DataValue")
    .tail(1)
)

plt.figure(figsize=(10,6))
sns.lineplot(data=avg_trends, x="YearStart", y="DataValue", hue="Topic", marker="o")
plt.title("Trends Over Time: Diabetes, Tobacco, and Obesity")
plt.ylabel("Average Prevalence (%)")

# Use .iloc[0] to grab single float values
plt.annotate(
    "Peak Diabetes Rate",
    xy=(max_diabetes["YearStart"].iloc[0], max_diabetes["DataValue"].iloc[0]),
    xytext=(2014, 55),
    arrowprops=dict(facecolor="black", arrowstyle="->")
)

plt.show()
```
